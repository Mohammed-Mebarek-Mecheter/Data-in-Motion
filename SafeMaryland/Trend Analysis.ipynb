```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("data/cleaned_MD_Crime_Data.csv")
df.head()
```

```python
# Group by 'Year' and calculate the mean or sum of overall crime rates per year
crime_rates = df.groupby('Year')['OverallCrimeRatePer100k'].mean()  # Adjust column name as per your dataset

# Plotting the trend
plt.figure(figsize=(12, 8))
plt.plot(crime_rates.index, crime_rates.values, marker='o', linestyle='-', color='b')
plt.title('Overall Crime Rate Trend in Maryland (1975-2020)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Overall Crime Rate per 100,000 people', fontsize=12)
plt.grid(True)
plt.xticks(range(min(crime_rates.index), max(crime_rates.index)+1, 5))  # Adjust x-axis ticks for better visibility
plt.tight_layout()
plt.show()
```

Overall crime rates in Maryland have decreased from 1975 to 2020. The crime rate in 1976 was 4,200 per 100,000 people, and it has decreased to 1,000 per 100,000 people in 2019, which represents a decrease of 47%.

The crime rate peaked in 1980 at 4,800 per 100,000 people. It then began a slow decline until 1984, when it reached a low of 3,800 per 100,000 people. The crime rate remained relatively stable until 1991, when it began to increase again. It reached a peak of 4,400 per 100,000 people in 1995.
