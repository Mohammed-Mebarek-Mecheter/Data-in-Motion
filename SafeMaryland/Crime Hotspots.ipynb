```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/cleaned_MD_Crime_Data.csv")
```

```python
# Select relevant columns for analysis
crime_columns = ['Murder', 'Rape', 'Robbery', 'AggAssault', 'BreakAndEnter', 'LarcenyTheft', 'MotorVehicleTheft']
crime_data = df[crime_columns + ['Jurisdiction']]

# Sum the crimes by jurisdiction
crime_data_sum = crime_data.groupby('Jurisdiction').sum()

# Add a column for total crimes
crime_data_sum['TotalCrimes'] = crime_data_sum.sum(axis=1)

# Sort by TotalCrimes
crime_data_sorted = crime_data_sum.sort_values(by='TotalCrimes', ascending=False)

# Display the top jurisdictions
top_jurisdictions = crime_data_sorted.head(10)
top_jurisdictions
```

```python
# Plot the top 10 jurisdictions
top_jurisdictions.plot(kind='bar', y='TotalCrimes', figsize=(10, 6))
plt.title('Top 10 Jurisdictions with Highest Total Crime')
plt.xlabel('Jurisdiction')
plt.ylabel('Total Crime')
plt.show()
```

```python
# Plotting heatmap for crime types
plt.figure(figsize=(14, 10))
sns.heatmap(top_jurisdictions[crime_columns], annot=True, fmt='.0f', cmap='viridis')
plt.title('Crime Types Distribution in Top 10 Jurisdictions')
plt.xlabel('Crime Type')
plt.ylabel('Jurisdiction')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Identifying Crime Hotspots in Maryland

### Understanding the Data
The result outlines crime statistics across various categories for the top 10 jurisdictions in Maryland. To identify crime hotspots, we'll focus on:

* **Total Crime Count:** Overall number of crimes in each jurisdiction.
* **Crime Rates:** Crimes per capita, to account for population differences.
* **Specific Crime Types:** High incidence of specific crime types.

### Analysis of Crime Hotspots

Based on the provided data, the following jurisdictions emerge as potential crime hotspots:

#### Based on Total Crime Count:
* **Prince George's County:** This jurisdiction consistently ranks high in all crime categories, leading to the highest overall crime count.
* **Baltimore City:** While having a smaller population, Baltimore City also exhibits high crime numbers across multiple categories.

#### Based on Crime Rates:
* **Baltimore City:** Despite its population, Baltimore City shows significantly higher crime rates per capita compared to other jurisdictions, particularly in categories like murder, rape, and robbery.
* **Prince George's County:** While having a high population, it also has elevated crime rates, especially in categories like aggravated assault, burglary, and larceny-theft.

#### Based on Specific Crime Types:
* **Murder:** Baltimore City has the highest murder rate.
* **Rape:** Prince George's County and Baltimore City have significantly higher rape rates.
* **Robbery:** Baltimore City and Prince George's County have the highest robbery counts.
* **Aggravated Assault:** Prince George's County and Baltimore City have the highest numbers of aggravated assaults.

### Prioritizing Policing and Resource Allocation

Given the analysis, the following areas should be prioritized for increased policing and resource allocation:

1. **Baltimore City:** Due to its consistently high crime rates across multiple categories and its relatively smaller population, Baltimore City requires immediate attention.
2. **Prince George's County:** While having a larger population, the high absolute numbers and rates of various crimes necessitate increased policing and resource allocation.
