```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/cleaned_MD_Crime_Data.csv")
df.head()
```

```python
# Select relevant columns for correlation analysis
columns_of_interest = ['Population', 'MurderPer100k', 'RapePer100k', 'RobberyPer100k',
                       'AggAssaultPer100k', 'BreakAndEnterPer100k', 'LarcenyTheftPer100k',
                       'MotorVehicleTheftPer100k']

crime_population_data = df[columns_of_interest]

# Calculate correlation matrix
correlation_matrix = crime_population_data.corr()

# Plotting the correlation matrix using seaborn heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix: Population vs Crime Rates')
plt.show()
```

The correlation matrix shows a moderate positive correlation between population size and crime rates, with a correlation coefficient of 0.37. This means that, in general, larger jurisdictions tend to have higher crime rates than smaller jurisdictions. However, the correlation is not very strong, so there are likely other factors that also play a role in crime rates.

The correlation coefficient for murder is 0.37, which is the same as the overall correlation. This suggests that population size has a similar effect on murder rates as it does on crime rates in general.

The correlation coefficient for rape is 0.22, which is lower than the overall correlation. This suggests that population size has a weaker effect on rape rates than it does on crime rates in general.

The correlation coefficient for robbery is 0.58, which is higher than the overall correlation. This suggests that population size has a stronger effect on robbery rates than it does on crime rates in general.

Overall, the correlation matrix shows that there is a moderate positive correlation between population size and crime rates in different jurisdictions. However, the strength of the correlation varies depending on the type of crime.
