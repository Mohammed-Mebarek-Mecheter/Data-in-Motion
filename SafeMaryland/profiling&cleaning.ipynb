```python
import numpy as np
# Define columns to treat for outliers
outlier_columns = ['MURDER', 'RAPE', 'ROBBERY', 'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT', 'GRAND TOTAL']

# Function to replace outliers with median and count changes
def replace_outliers_with_median(df, columns):
    changes_count = {}
    for column in columns:
        col_zscore = stats.zscore(df[column])
        abs_zscore = np.abs(col_zscore)
        median_value = df[column].median()
        # Identify rows to be changed
        rows_to_change = abs_zscore > 3
        # Count the number of changes
        changes_count[column] = rows_to_change.sum()
        # Replace outliers with median
        df[column] = np.where(rows_to_change, median_value, df[column])
    return df, changes_count

# Replace outliers in specified columns and get the count of changes
df_cleaned, changes_count = replace_outliers_with_median(df, outlier_columns)

# Print the number of rows changed for each column
for column, count in changes_count.items():
    print(f'{column}: {count} rows changed')
```
