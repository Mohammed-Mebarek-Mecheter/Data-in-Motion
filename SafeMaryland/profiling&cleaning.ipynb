```python
import numpy as np

# Handle zero values
df = df.replace(0, np.nan)
# Forward fill missing values
df = df.ffill()
```

```python
from sklearn.preprocessing import StandardScaler

def z_score_outliers(df, column, threshold=3):
    if column not in df.columns:
        print(f"Warning: Column '{column}' not found in the dataframe. Skipping.")
        return df
    if not pd.api.types.is_numeric_dtype(df[column]):
        print(f"Warning: Column '{column}' is not numeric. Skipping z-score outlier detection.")
        return df

    z_scores = stats.zscore(df[column].dropna())
    abs_z_scores = np.abs(z_scores)
    outliers = abs_z_scores > threshold

    # Replace outliers with NaN
    df.loc[df.index[outliers], column] = np.nan
    return df

def identify_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)].index

def drop_remaining_outliers(df, threshold_percent=5):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    all_outliers_index = set()
    outliers_per_column = {}

    for col in numeric_cols:
        outliers_index = identify_outliers(df, col)
        outliers_per_column[col] = len(outliers_index)
        all_outliers_index.update(outliers_index)

    # Calculate the percentage of rows that would be dropped
    drop_percentage = (len(all_outliers_index) / len(df)) * 100

    if drop_percentage > threshold_percent:
        print(f"Warning: Dropping {drop_percentage:.2f}% of rows exceeds the threshold of {threshold_percent}%.")
        print("Consider reviewing your data or adjusting the outlier criteria.")
        return df  # Return original dataframe without dropping

    print("Outliers per column:")
    for col, count in outliers_per_column.items():
        print(f"  {col}: {count} outliers")

    print(f"Total unique rows to be dropped: {len(all_outliers_index)}")
    df_dropped = df.drop(index=all_outliers_index)

    print(f"Original dataframe shape: {df.shape}")
    print(f"Cleaned dataframe shape: {df_dropped.shape}")
    print(f"Percentage of rows dropped: {drop_percentage:.2f}%")

    return df_dropped

# Define columns to process with z-score outlier detection
columns = [
    'VIOLENT CRIME TOTAL', 'PROPERTY CRIME TOTALS',
    'OVERALL CRIME RATE PER 100,000 PEOPLE',
    'VIOLENT CRIME RATE PER 100,000 PEOPLE',
    'PROPERTY CRIME RATE PER 100,000 PEOPLE',
    'MURDER PER 100,000 PEOPLE', 'RAPE PER 100,000 PEOPLE',
    'ROBBERY PER 100,000 PEOPLE', 'AGG. ASSAULT PER 100,000 PEOPLE',
    'B & E PER 100,000 PEOPLE', 'LARCENY THEFT PER 100,000 PEOPLE',
    'M/V THEFT PER 100,000 PEOPLE'
]

# Apply z-score outlier detection to specified columns
for column in columns:
    df = z_score_outliers(df, column)

# Replace NaN values with the median of the column
for column in columns:
    df[column] = df[column].fillna(df[column].median())

# Normalize/Standardize Data
scaler = StandardScaler()
df[columns] = scaler.fit_transform(df[columns])

# Remove duplicate rows
df = df.drop_duplicates()

# Check data types of columns
print(df.dtypes)

# Apply the outlier removal based on IQR method
df_cleaned = drop_remaining_outliers(df)

print("Outlier removal process is complete.")
```

```python
def convert_to_int64(df_cleaned):
    # List of columns that should logically be integers
    int_columns = [
        'Population', 'Murder', 'Rape', 'Robbery', 'AggAssault', 
        'BreakAndEnter', 'LarcenyTheft', 'MotorVehicleTheft', 
        'GrandTotal', 'ViolentCrimeTotal', 'PropertyCrimeTotal'
    ]
    
    for column in int_columns:
        if column in df_cleaned.columns:
            # Check if the column contains any NaN values
            if df_cleaned[column].isnull().any():
                print(f"Warning: Column '{column}' contains NaN values. These will be replaced with 0 before conversion.")
                df_cleaned[column] = df_cleaned[column].fillna(0)
            
            # Check if all values in the column are integers
            if np.all(np.equal(np.mod(df_cleaned[column], 1), 0)):
                # If all values are integers, convert directly to int64
                df_cleaned[column] = df_cleaned[column].astype('int64')
            else:
                # If there are float values, round before converting
                print(f"Warning: Column '{column}' contains float values. These will be rounded before conversion.")
                df_cleaned[column] = df_cleaned[column].round().astype('int64')
        else:
            print(f"Warning: Column '{column}' not found in the dataset. Skipping conversion to int64.")
    
    return df_cleaned
```
