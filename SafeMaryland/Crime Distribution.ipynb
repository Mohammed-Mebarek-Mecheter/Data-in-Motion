```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/cleaned_MD_Crime_Data.csv")
```

```python
# Select relevant columns for analysis
crime_types = ['Murder', 'Rape', 'Robbery', 'AggAssault', 'BreakAndEnter', 'LarcenyTheft', 'MotorVehicleTheft']
crime_data = df[crime_types + ['Year']]

# Sum the crimes by year
crime_data_sum = crime_data.groupby('Year').sum()

plt.figure(figsize=(14, 8))
sns.set(style="whitegrid") 

for crime_type in crime_types:
    plt.plot(crime_data_sum.index, crime_data_sum[crime_type], marker='o', label=crime_type, linewidth=2)

# Adding labels and title
plt.title('Crime Distribution in Maryland (1975-2020)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 8))
sns.set(style="whitegrid") 
crime_data_sum.plot.area(stacked=True, figsize=(14, 8))

plt.title('Crime Types in Maryland (1975-2020)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The distribution of different crime types in Maryland has changed over time. The most common type of crime in 1975 was larceny theft, followed by breaking and entering and motor vehicle theft. By 2020, the most common type of crime was larceny theft, followed by aggravated assault and robbery.

The number of larceny thefts and breaking and enterings has been steadily increasing since 1975. The number of motor vehicle thefts has also been increasing, but at a slower rate. The number of aggravated assaults and robberies has been more stable, with a slight increase in recent years.
